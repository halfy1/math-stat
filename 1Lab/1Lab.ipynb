```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

sizes = [10, 100, 1000]

distributions = {
    "normal": {
        "name_ru": "Нормальное N(0,1)",
        "sample": lambda n: np.random.normal(0, 1, n),
        "pdf": lambda x: stats.norm.pdf(x, 0, 1),
        "x_range": np.linspace(-4, 4, 300),
    },
    "cauchy": {
        "name_ru": "Коши C(0,1)",
        "sample": lambda n: np.random.standard_cauchy(n),
        "pdf": lambda x: stats.cauchy.pdf(x, 0, 1),
        "x_range": np.linspace(-4, 4, 300),
    },
    "laplace": {
        "name_ru": "Лапласа L(0, 1/√2)",
        "sample": lambda n: np.random.laplace(0, 1/np.sqrt(2), n),
        "pdf": lambda x: stats.laplace.pdf(x, 0, 1/np.sqrt(2)),
        "x_range": np.linspace(-4, 4, 300),
    },
    "poisson": {
        "name_ru": "Пуассона P(10)",
        "sample": lambda n: np.random.poisson(10, n),
        "pdf": lambda k: stats.poisson.pmf(k, 10),
        "x_range": np.arange(4, 18),
    },
    "uniform": {
        "name_ru": "Равномерное U(-√3, √3)",
        "sample": lambda n: np.random.uniform(-np.sqrt(3), np.sqrt(3), n),
        "pdf": lambda x: stats.uniform.pdf(x, -np.sqrt(3), 2*np.sqrt(3)),
        "x_range": np.linspace(-2.5, 2.5, 300),
    },
}

plt.rcParams['font.size'] = 10
plt.rcParams['axes.titlesize'] = 11
plt.rcParams['axes.labelsize'] = 9
plt.rcParams['figure.dpi'] = 100

def generate_histogram_plot(dist_key, dist_info):
    fig, axes = plt.subplots(1, 3, figsize=(12, 3.5))
    fig.suptitle(f"Распределение: {dist_info['name_ru']}", 
                 fontsize=13, y=1.02, weight='bold')
    
    for col, n in enumerate(sizes):
        ax = axes[col]
        
        sample = dist_info["sample"](n)
        x = dist_info["x_range"]
        
        if dist_key == "cauchy":
            plot_sample = sample[np.abs(sample) < 10]
        else:
            plot_sample = sample
        
        ax.hist(plot_sample, bins="auto", density=True,
                alpha=0.65, color="steelblue", edgecolor="black", 
                linewidth=0.5, label="Гистограмма")
        
        ax.plot(x, dist_info["pdf"](x), 
                color="red", linewidth=2.5, label="Теоретическая плотность")
        
        ax.set_title(f"n = {n}", fontsize=10)
        ax.set_xlabel("x", fontsize=9)
        ax.set_ylabel("Плотность", fontsize=9)
        ax.legend(fontsize=8, loc="best")
        ax.grid(alpha=0.3, linestyle="--", linewidth=0.5)
    
    plt.tight_layout()
    
    filename = f"lab1_{dist_key}.png"
    plt.savefig(filename, dpi=200, bbox_inches="tight")
    plt.close()
    

if __name__ == "__main__":
    
    for dist_key, dist_info in distributions.items():
        generate_histogram_plot(dist_key, dist_info)
```
